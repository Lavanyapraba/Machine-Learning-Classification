```python
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True,dtype=int)
```

```python
dataset
```

```python
dataset[["Purchased"]].value_counts()
```

```python
dataset.columns
```

```python
Independent=dataset[[ 'Age', 'EstimatedSalary','Gender_Male']]
```

```python
Independent
```

```python
Independent.shape
```

```python
Dependent=dataset[["Purchased"]]
```

```python
Dependent
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
x_train,x_test,y_train,y_test=train_test_split(Independent,Dependent,test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
param_grid = {
    'n_neighbors': [ 5, 10,50,100],  
    'weights': ['uniform', 'distance',None],  
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'] } 
grid=GridSearchCV(KNeighborsClassifier(),param_grid,cv=5,refit=True,n_jobs=-1,verbose=3)
```

```python
grid.fit(x_train,y_train)
```

```python
y_pred=grid.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,roc_auc_score
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
print(cm)
```

```python
clf_rpt=classification_report(y_test,y_pred)
```

```python
print (clf_rpt)
```

```python
roc_score=roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
```

```python
roc_score
```

```python
re=grid.cv_results_
```

```python
print ("Best parameters are :", grid.best_params_)
```
