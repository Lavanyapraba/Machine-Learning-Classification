```python
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True,dtype=int)
```

```python
dataset
```

```python
dataset.columns
```

```python
Independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
```

```python
Independent
```

```python
Independent.shape
```

```python
Dependent=dataset[['classification_yes']]
```

```python
Dependent
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
Independent=sc.fit_transform(Independent)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
param_grid = {
    'n_neighbors': [ 5, 10,50,100],  
    'weights': ['uniform', 'distance',None],  
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'] } 
grid=GridSearchCV(KNeighborsClassifier(),param_grid,verbose=3,refit=True,cv=5,n_jobs=-1)
```

```python
grid.fit(Independent,Dependent)
```

```python
grid_pred=grid.predict(Independent)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Dependent,grid_pred)
```

```python
print (cm)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(Dependent,grid_pred)
```

```python
print(clf_report)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(Dependent,grid.predict_proba(Independent)[:,1])
```

```python
re=grid.cv_results_
```
