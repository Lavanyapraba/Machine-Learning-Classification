```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset["classification_yes"].value_counts()
```

```python
dataset.columns
```

```python
Independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
```

```python
Independent.shape
```

```python
Independent
```

```python
Dependent=dataset[[ 'classification_yes']]
```

```python
Dependent
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
Independent=sc.fit_transform(Independent)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
param_grid={'kernel':['linear','poly','rbf','sigmoid'],'C':[10,100,1000,2000,3000],'gamma':['scale','auto']}

grid=GridSearchCV(SVC(probability=True),param_grid,refit=True,verbose=3,n_jobs=-1,cv=5)

grid.fit(Independent,Dependent)
```

```python
grid_pred=grid.predict(Independent)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Dependent,grid_pred)
```

```python
print (cm)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(Dependent,grid_pred)
```

```python
print(clf_report)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(Dependent,grid.predict_proba(Independent)[:,1])
```

```python
re=grid.cv_results_
```

```python
print("Best parameters found: ", grid.best_params_)
```

```python
table=pd.DataFrame.from_dict(re)
```
