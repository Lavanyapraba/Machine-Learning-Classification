```python
import pandas as pd
import numpy as ny
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True,dtype=int)
```

```python
dataset
```

```python
dataset.columns
```

```python
Independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
```

```python
Independent
```

```python
Independent.shape
```

```python
Dependent=dataset[['classification_yes']]
```

```python
Dependent
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import ComplementNB
param_grid={'alpha':[1,10,100],'force_alpha':[True,False],'fit_prior':[True,False],'norm':[False]}
grid=GridSearchCV(ComplementNB(),param_grid,verbose=True,cv=5,refit=True,n_jobs=-1,scoring='f1')
```

```python
grid.fit(Independent,Dependent)
```

```python
grid_pred=grid.predict(Independent)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,roc_auc_score
```

```python
cm=confusion_matrix(Dependent,grid_pred)
```

```python
print(cm)
```

```python
clf_rpt=classification_report(Dependent,grid_pred)
```

```python
print(clf_rpt)
```

```python
roc_score=roc_auc_score(Dependent,grid.predict_proba(Independent)[:,1])
```

```python
roc_score
```

```python
re=grid.cv_results_
```

```python
print("Best parameters found: ", grid.best_params_)
```

```python
table=pd.DataFrame.from_dict(re)
```
